### <a name="anchor01" />Présentation d'EcoBalyze

<div style="padding-right: 200px;">

---    
    
__Écobalyse__ est un outil développé par l'État français pour calculer `l'impact écologique` des `produits textiles` et `alimentaires` distribués en `France`.  
Il vise à fournir des informations sur `l'empreinte environnementale` de ces produits, permettant ainsi aux consommateurs de prendre des décisions plus éclairées et durables sur leurs choix de consommation.

Visitez le site d'__Écobalyse__, [ici](https://ecobalyse.beta.gouv.fr/)<sup>1 , 2 , 3 , 4 , 5 , 6 , 7 , 8</sup>. 

&nbsp;    

__A voir également :__

- [ADEME : Communiquer sur la performance environnementale : approche semi-spécifique](https://agirpourlatransition.ademe.fr/entreprises/ecoconception/communiquer-performance)  
- [ADEME : Base Empreinte](https://base-empreinte.ademe.fr/)

    
<i>En savoir plus</i> : <sup>1</sup>[Simulateur](https://ecobalyse.beta.gouv.fr/#/textile/simulator) , <sup>2</sup>[Explorateur](https://ecobalyse.beta.gouv.fr/#/explore/textile) , <sup>3</sup>[API](https://ecobalyse.beta.gouv.fr/#/api) , <sup>4</sup>[Documentation](https://fabrique-numerique.gitbook.io/ecobalyse) , <sup>5</sup>[GitHub](https://github.com/MTES-MCT/ecobalyse) , <sup>6</sup>[fashiongreenhub.org](https://www.fashiongreenhub.org/2023/08/31/lecobalyse-un-outil-pour-accelerer-la-mise-en-place-de-laffichage-environnemental/) , <sup>7</sup>[gorfou.fr](https://www.gorfou.fr/ecobalyse-calcule-vos-impacts-environnementaux/) , <sup>8</sup>[mtes-mct.github.io](https://mtes-mct.github.io/portail/startup/ecobalyse/)

#### <a name="anchor01_A" />Impact Environnemental des Textiles

<div style="padding-right: 200px;">

---    

L'industrie textile est l'une des plus polluantes au monde<sup>1 , 2 , 3 , 4 , 5</sup>.    
    
En lien avec les préoccupations actuelles, __Écobalyse__ vise à accélérer la mise en place de `l'affichage environnemental`, pour favoriser un `modèle de production` plus `durable`.

</div>

##### <a name="anchor01_A1" />Points Clefs

<div style="padding-right: 200px;">

__Voici quelques points clés à propos d'Écobalyse__ :
    
- `Objectif` : __Écobalyse__ permet de comprendre et de calculer les impacts écologiques des produits distribués en France.

- `Éco-score` : __Écobalyse__ propose un éco-score pour informer les consommateurs sur l'impact environnemental des produits qu'ils achètent.    
    
- `Collaboration ouverte` : __Écobalyse__ est un mode de collaboration ouvert à la critique et aux suggestions, dans le but d'aider à élaborer la future méthode réglementaire française.

- [`API ouverte`](https://api.gouv.fr/les-api/api-ecobalyse) : __Écobalyse__ propose une __interface de programmation applicative__ (__API__) pour connecter le calculateur Écobalyse à d'autres services numériques. L'[__API__](https://api.gouv.fr/les-api/api-ecobalyse) est testée dans le cadre de l'expérimentation Xtex, conformément à la loi Climat et Résilience.
    
    Ci-après quelques exemples d'utilisation de l'[__API__](https://api.gouv.fr/les-api/api-ecobalyse) __Écobalyse__ pour estimer les impacts environnementaux des produits textiles : 
    
        1. Interroger la base de données des indicateurs d'impacts environnementaux des produits textiles. 
        Récupérer des informations sur les impacts environnementaux d'un vêtement en fonction de critères tels que 
        la traçabilité, la matière et le recyclage.
    
        2. Calculer les impacts pour chaque produit.
        Estimer, pour chaque produit textile : l'impact carbone, l'impact sur la couche d'ozone, l'impact sur la 
        qualité de l'eau, le coût énergétique, l'impact sur l'eutrophisation de l'eau et des terres, l'impact sur 
        l'acidification terrestre des eaux douces, l'utilisation du sol, l'utilisation de minéraux. 
    
        3. Tester le simulateur.
        Explorer les impacts environnementaux de différents produits textiles.
    
Toutes marques, producteurs, ou distributeurs peuvent contribuer à améliorer le calcul d'impacts écologiques, en partageant leurs données et en participant aux travaux collectifs.

<i>En savoir plus</i> : <sup>1</sup>[la-goose.com](https://la-goose.com/les-impacts-environnementaux-du-textile-comment-les-reduire/) , <sup>2</sup>[oxfamfrance.org](https://www.oxfamfrance.org/agir-oxfam/impact-de-la-mode-consequences-sociales-environnementales/) , <sup>3</sup>[ecologie.gouv.fr](https://www.ecologie.gouv.fr/mieux-informer-consommateur-vers-affichage-environnemental-des-vetements-indiquer-leur-impact) , <sup>4</sup>[climateseed.com](https://climateseed.com/fr/blog/secteur-du-textile-impact-environnemental-et-r%C3%A9glementation) , <sup>5</sup>[wwf.ch](https://www.wwf.ch/fr/nos-objectifs/rapport-du-wwf-sur-lindustrie-de-lhabillement-et-des-textiles)

<div style="padding-right: 200px;">

__Le poids d’un T-shirt peut varier en fonction de sa taille, du type de tissu et de la conception__.  

_Dans l’industrie textile, le poids des T-shirts est généralement mesuré en onces par yard carré (oz/yd²) ou en grammes par mètre carré (gsm)_. 

En moyenne, un T-shirt peut peser entre 100 et 300 grammes<sup>1</sup>. Cela se traduit par 0,1 à 0,3 kilogrammes<sup>1</sup>.

Cependant, en termes de densité du tissu (souvent mesurée en onces par yard carré (oz/yd²) ou en grammes par mètre carré (gsm)), les T-shirts varient généralement de léger, environ 4 onces (environ 135 gsm), à lourd, plus de 6 onces (environ 200 gsm)<sup>2 , 3</sup>.  
    
Les __T-shirts légers__ sont appréciés pour leur respirabilité et leur confort, ce qui les rend idéaux pour une utilisation décontractée ou dans des climats plus chauds<sup>2</sup>.  

D’autre part, les __T-shirts lourds__ offrent plus de durabilité et d’opacité, ce qui est particulièrement bénéfique pour les marques premium, les vêtements de travail, ou pour ceux qui recherchent une sensation plus substantielle<sup>2</sup>.

Il est important de noter que ces poids sont des moyennes et peuvent varier en fonction de la marque et du style spécifiques du T-shirt.

En savoir plus : <sup>1</sup>[measuringhow.com](https://www.measuringhow.com/) , <sup>2</sup>[printkk.com](https://www.printkk.com) , <sup>3</sup>[tshirtgrowth.com](https://www.tshirtgrowth.com) , <sup>4</sup>[silverbobbin.com](https://www.silverbobbin.com) , <sup>5</sup>[noclout.com](https://www.noclout.com)
    
&nbsp;

__Voici une liste standardisée des poids de T-shirts__:

- __T-shirts légers__ : Autour de 4 onces (environ 135 gsm). Ces T-shirts sont appréciés pour leur respirabilité et leur confort, ce qui les rend idéaux pour une utilisation décontractée ou dans des climats plus chauds<sup>1 , 2</sup>.  
    
- __T-shirts de poids standard__ : De 4 à 6 onces (environ 135 à 200 gsm). Ces T-shirts offrent un équilibre entre durabilité et confort, ce qui les rend adaptés pour une utilisation quotidienne<sup>1 , 2</sup>.  
    
- __T-shirts lourds__ : Plus de 6 onces (plus de 200 gsm). Ces T-shirts offrent plus de durabilité et d’opacité, ce qui est particulièrement bénéfique pour les marques premium, les vêtements de travail, ou pour ceux qui recherchent une sensation plus substantielle<sup>1 , 2</sup>.  

Il est important de noter que ces poids sont des moyennes et peuvent varier en fonction de la marque et du style spécifiques du T-shirt.  
    
De plus, le poids d’un T-shirt peut également être influencé par d’autres facteurs tels que la taille du T-shirt, le type de tissu utilisé, et la conception du T-shirt<sup>3</sup>.

En savoir plus : <sup>1</sup>[printkk.com](https://www.printkk.com) , <sup>2</sup>[tshirtgrowth.com](https://www.tshirtgrowth.com) , <sup>3</sup>[printerpreneur.com](https://www.printerpreneur.com) , <sup>4</sup>[zegaapparel.com](https://www.zegaapparel.com) , <sup>5</sup>[printful.com](https://www.printful.com)

##### <a name="anchor02_A2" />Fonction de récupération des données

````python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Définition de la fonction de Scraping / Selenium
def get_explorer_url(url, columns):
    """
    A documenter ultérieurement ...
    """
    # Créer une nouvelle instance de navigateur
    driver = webdriver.Chrome()
    
    # Rechercher la page à afficher 
    driver.get(url)
    
    # Attendre que la page se charge
    wait = WebDriverWait(driver, 10)
    
    # Extraire les données find_elements(By.CSS_SELECTOR, 'table tr')
    # Sélecteur CSS pour les lignes du tableau
    data = []
    rows = driver.find_elements(By.CSS_SELECTOR, 'table tr') 
    
    # Sélecteur CSS pour les colonnes du tableau
    for row in rows:
        cols = row.find_elements(By.CSS_SELECTOR, 'td')  
        cols = [col.text for col in cols]
        data.append(cols)
    
    # Fermer le webdriver
    driver.close()

    # Créer le dataframe
    df = pd.DataFrame(data)
    
    # Nommer les colonnes
    df.columns = columns
    
    return df
````

````python
import requests
import pandas as pd

link = 'https://ecobalyse.beta.gouv.fr/api/'

# Composition du Tshirt synthétique (150g) - Asie - Mode ultra fast fashion

# T-shirt / Polo : 0,150kg, matière : 95% Polyester provenance  par défaut (Asie - Pacifique), 
# 5% Polypropylène provenance  par défaut (Europe), filature : Région - Asie, 
# Tricotage moyen (par défaut) : Région - Asie, ennoblissement : Région - Asie, impression : non, 
# confection : Région - Asie (100 de transport aérien), distribution : France, utilisation : France, 
# fin de vie : France

# {
#   "mass": 0.15,
#   "materials": [
#     {
#       "id": "ei-pet",
#       "share": 0.95
#     },
#     {
#       "id": "ei-pp",
#       "share": 0.05
#     }
#   ],
#   "product": "tshirt",
#   "countrySpinning": "RAS",
#   "countryFabric": "RAS",
#   "countryDyeing": "RAS",
#   "countryMaking": "RAS",
#   "airTransportRatio": 1,
#   "fabricProcess": "knitting-mix",
#   "business": "large-business-without-services",
#   "marketingDuration": 65,
#   "numberOfReferences": 100000,
#   "price": 10,
#   "traceability": false
# }

data = {
        # Masse du produit fini, exprimée en kilogrammes. Min 0.01 
        'mass': 0.15,
        
        # Liste des matières composant le produit (voir : textile/materials).
        # Le format de chaque entrée est composé de l'identifiant de la matière 
        # et de la part du produit qu'elle représente, optionnellement du procédé 
        # de filature ou filage pour la matière (ConventionalSpinning, UnconventionalSpinning, 
        # SyntheticSpinning), et optionnellement du code du pays d'origine de la matière.
        'materials': [
                        {'id': 'ei-pet', 'share': 0.95},
                        {'id': 'ei-pp', 'share': 0.05}
                     ],
                        
        # Identifiant du produit (voir : textile/products)
        'product': 'tshirt',
    
        # Code pays pour l'étape de Filature (voir: textile/countries) 
        # Si non spécifié, le pays de filature pris en considération est 
        # celui de production de la matière la plus représentée dans le mix.
        'countrySpinning': 'RAS',
        
        # Code pays pour l'étape de Tissage/Tricotage (voir : textile/countries) 
        'countryFabric': 'RAS',
    
        # Code pays pour l'étape de Teinture (voir : textile/countries)
        'countryDyeing': 'RAS',
    
        # Code pays pour l'étape de Confection (voir : textile/countries)
        'countryMaking': 'RAS',
    
        # Part de transport aérien entre l'étape de Confection et l'étape de Distribution, entre 0 et 1 
        'airTransportRatio': 1,
    
        # Le processus utilisé pour tisser ou tricoter le tissu, parmi :
        # knitting-mix ┃ knitting-fully-fashioned ┃ knitting-integral 
        # ┃ knitting-circular ┃ knitting-straight ┃ weaving
        'fabricProcess': 'knitting-mix',
    
        # Type d'entreprise et d'offre de services, parmi :
        # small-business ┃ large-business-with-services ┃ large-business-without-services
        'business' : 'large-business-without-services',
    
        # Durée de commercialisation, en jours. Min 1 ┃ Max 999
        'marketingDuration' : 65,
    
        # Nombre de références au catalogue de la marque. Min 1 ┃ Max 999999 
        'numberOfReferences' : 100000,
    
        # Prix du produit, en Euros (€). Min 1 ┃ Max 1000
        'price' : 10,
    
        # Traçabilité renforcée. true ┃ false
        'traceability' : 'false'
       }

# Scrapper API avec le dictionnaire
result2 = requests.post(link + 'textile/simulator', json=data)
    
# Vérifiez le statut de la requête
if result2.status_code == 200:
    # Convertir la réponse en JSON
    response_json = result2.json()
    
    # Extraire la liste des impacts environnementaux dans un DataFrame
    df_impacts = pd.DataFrame(response_json['impacts'].items(), columns=['Nom', 'Tshirt synthétique (150g) - Asie - Mode ultra fast fashion'])
    print(response_json['description'] + '\n')
    
else:
    print('La requête a échoué avec le code de statut :', result2.status_code)

# Transposition du DataFrame
df_transpose = df_impacts.transpose()

df_transpose.columns = df_transpose.iloc[0]  # Faire de la première ligne les libellés des colonnes
df_transpose = df_transpose[1:]  # Supprimer la première ligne

# Affichage du DataFrame
print('Nbre de ligne(s) du tableau résultat : ' + str(df_transpose.shape[0]) + '\n')
df_transpose
````

###### <a name="anchor02_C1-42" /><u>Exemple</u> : Liste de T-shirts depuis un fichier CSV

````python
import pandas as pd
import numpy as np

import requests
from requests.exceptions import RequestException

# Lire le fichier CSV dans un DataFrame
df = pd.read_csv('./api_t-shirt_data.csv', encoding="ISO-8859-1", sep=';')

# Initialiser un DataFrame vide pour stocker les résultats
df_results = pd.DataFrame()

link = 'https://ecobalyse.beta.gouv.fr/api/'

try:
    # Bloc de code à essayer d'exécuter
    
    for index, row in df.iterrows():
        # Mettre à jour les valeurs de data avec les valeurs de la ligne actuelle
        data = {
            'mass': row['mass'],
            'materials': [
                            {'id': row['material_id1'], 'share': row['share1']},
                            {'id': row['material_id2'], 'share': row['share2']},
                            {'id': row['material_id3'], 'share': row['share3']}                
                         ],
            'product': row['product'],
            'countrySpinning': row['countrySpinning'],
            'countryFabric': row['countryFabric'],
            'countryDyeing': row['countryDyeing'],
            'countryMaking': row['countryMaking'],
            'airTransportRatio': row['airTransportRatio'],
            'fabricProcess': row['fabricProcess'],
            'business' : row['business'],
            'marketingDuration' : row['marketingDuration'],
            'numberOfReferences' : row['numberOfReferences'],
            'price' : row['price'],
            'traceability' : row['traceability']
        }

        # Scrapper API avec le dictionnaire
        result2 = requests.post(link + 'textile/simulator', json=data)

        # Vérifiez le statut de la requête
        result2.raise_for_status()
    
        if result2.status_code == 200:
            # Convertir la réponse en JSON
            response_json = result2.json()
        
            # Extraire la liste des impacts environnementaux dans un DataFrame
            df_impacts = pd.DataFrame(response_json['impacts'].items(), columns=['Nom', row['description']])
            df_impacts = df_impacts.fillna('')
        
            # Transposition du DataFrame
            df_transpose = df_impacts.transpose()
            df_transpose.columns = df_transpose.iloc[0]  # Faire de la première ligne les libellés des colonnes
            df_transpose = df_transpose[1:]  # Supprimer la première ligne
        
            # Ajouter le DataFrame transposé aux résultats
            df_results = pd.concat([df_results, df_transpose])

        else:
            # print('La requête a échouée avec le code de statut :' + str(result2.status_code))
            raise RequestException('Erreur')

except RequestException as e: 
    # Bloc de code à exécuter en cas d'erreur
        
    print('La requête a échouée avec le code de statut : ' + str(result2.status_code))
    print(f"Plus précisément : : {e}" + '\n')

else:
    # Bloc de code à exécuter si aucune erreur ne s'est produite
    
    # Titre
    print('Liste des impacts de T-shirts depuis un fichier CSV (seules les 10 dernières lignes du résultat sont affichées).' + '\n')

    # Arrondir les valeurs des colonnes 'ecs' et 'pef' à l'entier supérieur
    df_results['ecs'] = np.ceil(df_results['ecs'])
    df_results['pef'] = np.ceil(df_results['pef'])
    
    # Transformer le DataFrame en fichier JSON
    df_results.to_json('./1-RECOLTE-DATA/API/f_api_t-shirt_data.json', orient='split', force_ascii=False)  

    # Sauvegarder le DataFrame en fichier CSV
    df_results.to_csv('./1-RECOLTE-DATA/API/f_api_t-shirt_data.csv', encoding='ISO-8859-1', index=True)
        
    # Affichage du DataFrame
    print('Nbre de ligne(s) totale(s) du tableau résultat : ' + str(df_results.shape[0]) + '\n')

    # Affichage du DataFrame
    print(df_results.tail(10))
       
````

###### <a name="anchor02_C1-43" /><u>Exemple</u> : Génération de T-Shirts avec des données aléatoires 

<div style="padding-right: 200px;">

<b>mass : </b> masse du produit fini, exprimée en kilogrammes. Min 0.10 , Max 0.30 , par pas de 0.1  
<b>countrySpinning : </b> code pays pour l'étape de Filature. Si non spécifié, le pays de filature pris en considération est celui de production de la matière la plus représentée dans le mix.

</div>

````python
import numpy as np

# mass : masse du produit fini, exprimée en kilogrammes. Min 0.10 , Max 0.30 , par pas de 0.1
def get_random_mass():
    # Générer une valeur de masse aléatoire entre 0.1 et 0.3 avec un pas de 0.1
    mass = np.random.choice(np.arange(0.1, 0.4, 0.1))
    return mass
````

````python
import numpy as np

# Associer un code pays pour les étapes de fabrication (filature, tissage / tricotage, teinture , confection)
def get_random_country():
    # Liste des valeurs possibles de pays
    country_values = ['BD', 'KH', 'CN', 'FR', '---', 'IN', 'MA', 'MM', 'PK', 'RAF', 'RLA', \
                      'RNA', 'RAS', 'REE', 'REO', 'RME', 'ROC', 'TN', 'TR', 'VN']
    # Générérer une valeur aléatoire de code pays pour le T-Shirt
    country_code = np.random.choice(country_values)
    return country_code
````

````python
import numpy as np

# Générer une valeur aléatoire pour la part de transport aérien (0%, 33%, ou 100%)
def get_random_airTransportRatio():
    # Liste des valeurs possibles de airTransportRatio 
    airTransportRatio_values = [0, 0.33, 1]
    # Générer une valeur aléatoire
    airTransportRatio = np.random.choice(airTransportRatio_values)
    return airTransportRatio
````

````python
import numpy as np

# Générer une valeur aléatoire de prix, entre 10€ et 40€ maxi pour un T-Shirt.
def get_random_price_tshirt():
    # Définir la valeur initiale du prix d'un T-Shirt : 10€
    initial_price = 10    
    # Calculer le pourcentage maximum possible pour que le prix final ne dépasse pas 40€
    max_percentage = (40 - initial_price) / initial_price    
    # Générer un pourcentage aléatoire entre 0% et le pourcentage maximum
    random_percentage = np.random.uniform(0, max_percentage)    
    # Calculer le prix final en appliquant le pourcentage aléatoire
    final_price = initial_price * (1 + random_percentage)
    # Retourner le prix final avec 2 chiffres après la virgule
    return round(final_price, 2)
````

````python
import requests

import pandas as pd
import numpy as np

link = 'https://ecobalyse.beta.gouv.fr/api/'

# Composition aléatoire de paramètres du Tshirt
mass = get_random_mass()
countrySpinning = get_random_country()
countryFabric = get_random_country()
countryDyeing = get_random_country()
countryMaking = get_random_country()
airTransportRatio = get_random_airTransportRatio()
price = get_random_price_tshirt()

# Générer un résumé du nom du T-Shirt sur la base des valeurs aléatoires
nomTshirt = 'Tshirt (' + str(int(mass*1000)) + 'g) - ' + countryMaking + ' - ' \
            + str(int(airTransportRatio*100)) + '%' + ' - ' + str(price) + '€.'

data = {
        'mass': mass,
        
        # Liste des matières composant le produit (voir : textile/materials).
        # Le format de chaque entrée est composé de l'identifiant de la matière 
        # et de la part du produit qu'elle représente, optionnellement du procédé 
        # de filature ou filage pour la matière (ConventionalSpinning, UnconventionalSpinning, 
        # SyntheticSpinning), et optionnellement du code du pays d'origine de la matière.
        'materials': [
                        {'id': 'ei-pet', 'share': 0.95},
                        {'id': 'ei-pp', 'share': 0.05}
                     ],
                        
        # Identifiant du produit (voir : textile/products)
        'product': 'tshirt',
    
        # Code pays pour l'étape de Filature (voir: textile/countries) 
        # Si non spécifié, le pays de filature pris en considération est 
        # celui de production de la matière la plus représentée dans le mix.
        'countrySpinning': countrySpinning,
        
        # Code pays pour l'étape de Tissage/Tricotage (voir : textile/countries) 
        'countryFabric': countryFabric,
    
        # Code pays pour l'étape de Teinture (voir : textile/countries)
        'countryDyeing': countryDyeing,
    
        # Code pays pour l'étape de Confection (voir : textile/countries)
        'countryMaking': countryMaking,
    
        # Part de transport aérien entre l'étape de Confection et l'étape de Distribution, entre 0 et 1 
        'airTransportRatio': airTransportRatio,
    
        # Le processus utilisé pour tisser ou tricoter le tissu, parmi :
        # knitting-mix ┃ knitting-fully-fashioned ┃ knitting-integral 
        # ┃ knitting-circular ┃ knitting-straight ┃ weaving
        'fabricProcess': 'knitting-mix',
    
        # Type d'entreprise et d'offre de services, parmi :
        # small-business ┃ large-business-with-services ┃ large-business-without-services
        'business' : 'large-business-without-services',
    
        # Durée de commercialisation, en jours. Min 1 ┃ Max 999
        'marketingDuration' : 65,
    
        # Nombre de références au catalogue de la marque. Min 1 ┃ Max 999999 
        'numberOfReferences' : 100000,
    
        # Prix du produit, en Euros (€). Min 1 ┃ Max 1000
        'price' : price,
    
        # Traçabilité renforcée. true ┃ false
        'traceability' : 'false'
       }

# Scrapper API avec le dictionnaire
result2 = requests.post(link + 'textile/simulator', json=data)
    
# Vérifiez le statut de la requête
if result2.status_code == 200:
    # Convertir la réponse en JSON
    response_json = result2.json()
    
    # Extraire la liste des impacts environnementaux dans un DataFrame
    df_impacts = pd.DataFrame(response_json['impacts'].items(), columns=['Nom', nomTshirt])
    print(response_json['description'] + '\n')
    
else:
    print('La requête a échoué avec le code de statut :', result2.status_code)

# Transposition du DataFrame
df_transpose = df_impacts.transpose()

df_transpose.columns = df_transpose.iloc[0]  # Faire de la première ligne les libellés des colonnes
df_transpose = df_transpose[1:]  # Supprimer la première ligne

# Arrondir les valeurs des colonnes 'ecs' et 'pef' à l'entier supérieur
df_transpose['ecs'] = np.ceil(df_transpose['ecs'])
df_transpose['pef'] = np.ceil(df_transpose['pef'])

# Affichage du DataFrame
print('Nbre de ligne(s) du tableau résultat : ' + str(df_transpose.shape[0]) + '\n')
df_transpose
````

###### <a name="anchor02_C1-5" />Calcul d'un impact environnemental spécifique d'un produit textile

<b>Default:</b> cch  
<b>Allowed:</b> acd ┃ bvi ┃ cch ┃ <b>ecs</b> ┃ etf ┃ fru ┃ fwe ┃ htc ┃ htn ┃ ior ┃ ldu ┃ mru ┃ ozd ┃ pco ┃ <b>pef</b> ┃ pma ┃ swe ┃ tre ┃ wtu

<b>acd:</b> Acidification (mol éq. H+)  
<b>bvi:</b> Biodiversité locale (BVI)  
<b>cch:</b> Changement climatique (kg éq. CO2)  
<b><u>ecs:</u></b> Score d'impacts (µPts d'impact)  
<b>etf:</b> Écotoxicité de l'eau douce (CTUe)  
<b>fru:</b> Utilisation de ressources fossiles (MJ)  
<b>fwe:</b> Eutrophisation eaux douces (kg éq. P)  
<b>htc:</b> Toxicité humaine - cancer (CTUh)  
<b>htn:</b> Toxicité humaine - non-cancer (CTUh)  
<b>ior:</b> Radiations ionisantes (éq. kBq U235)  
<b>ldu:</b> Utilisation des sols (sans dimension (pt))  
<b>mru:</b> Utilisation de ressources minérales et métalliques (kg éq. Sb)  
<b>ozd:</b> Appauvrissement de la couche d'ozone (kg éq. CFC 11)  
<b>pco:</b> Formation d'ozone photochimique (kg éq. COVNM)  
<b><u>pef:</u></b> Score PEF (µPt)  
<b>pma:</b> Particules (incidence de maladie)  
<b>swe:</b> Eutrophisation marine (kg éq. N)  
<b>tre:</b> Eutrophisation terrestre (mol éq. N)  
<b>wtu:</b> Utilisation de ressources en eau (m³)
